```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from rapidfuzz import process
import pickle
```

```python
# Cargar el archivo
ruta_archivo = r"D:\Maestria\CLASES\TFE\DATOS_LIGA_COLOMBIANA.xlsx"
```

```python
# Variable data set
df = pd.read_excel(ruta_archivo)
```

```python
# Mostrar los datos iniciales
df.head(3000)
```

```python
# Revisar estructura inicial
print(df.shape)       # filas y columnas
print(df.columns)     # nombres de columnas
```

```python
# Estandarizar nombres de columnas (minúsculas y sin espacios)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# Identificar tipos de variables
print(df.dtypes)
```

```python
# Detectar valores nulos
print(df.isnull().sum())
print(df[df.isnull().any(axis=1)])
```

```python
# Eliminar registros que contienen al menos un valor nulo
df = df.dropna()
```

```python
# Mostrar cantidad nueva de registros
print(df.shape)  
```

```python
# Identificar registros duplicados (duplicados completos)
duplicados = df[df.duplicated(keep=False)]

# Mostrar los registros duplicados
print("Registros duplicados encontrados:")
print(duplicados)
```

```python
# Estandarizar categorías
df['equipo_local'] = df['equipo_local'].str.strip().str.lower()
df['equipo_visitante'] = df['equipo_visitante'].str.strip().str.lower()
```

```python
with open('df_guardado.pkl', 'wb') as f:
    pickle.dump(df, f)
```
